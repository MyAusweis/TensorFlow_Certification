```
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow_text as text
```

```
ds = tfds.load('imdb_reviews', split='train')
docs = []
for ex in ds.take(4):
  print(ex['text'])
  docs.append(ex['text'])
docs
```

```
tokenizer = text.WhitespaceTokenizer()
tokens = tokenizer.tokenize(docs)
tokens.to_list()
```

```
tokenizer = text.UnicodeScriptTokenizer()
tokens = tokenizer.tokenize(docs)
tokens.to_list()
```

```
short_docs = tf.data.Dataset.from_tensor_slices([['Never tell me the odds.'], ["It's a trap!"]])
tokenizer = text.UnicodeScriptTokenizer()
tokenized_docs = short_docs.map(lambda x: tokenizer.tokenize(x))
iterator = iter(tokenized_docs)
print(next(iterator).to_list())
print(next(iterator).to_list())
```

```
tokenizer = tfds.features.text.Tokenizer()

vocabulary_set = set()
for doc in docs:
  some_tokens = tokenizer.tokenize(doc.numpy())
  vocabulary_set.update(some_tokens)

vocab_size = len(vocabulary_set)
vocab_size
```

```
encoder = tfds.features.text.TokenTextEncoder(vocabulary_set)
```

```
sample_doc = docs[0].numpy()
print(sample_doc)
','.join([str(a) for a in encoder.encode(sample_doc)])
```

```
tokenizer = text.WhitespaceTokenizer()
tokens = tokenizer.tokenize(docs)

# Is capitalized?
f1 = text.wordshape(tokens, text.WordShape.HAS_TITLE_CASE) 
# Are all letters uppercased?
f2 = text.wordshape(tokens, text.WordShape.IS_UPPERCASE)
# Does the token contain punctuation?
f3 = text.wordshape(tokens, text.WordShape.HAS_SOME_PUNCT_OR_SYMBOL)
# Is the token a number?
f4 = text.wordshape(tokens, text.WordShape.IS_NUMERIC_VALUE)

print(f1)
print(f2.to_list())
print(f3.to_list())
print(f4.to_list())
```

```
def extract_shape(doc, shape):
  return [token.numpy() for token in doc if text.wordshape(token, shape)]

title_case = extract_shape(tokens[0], text.WordShape.HAS_TITLE_CASE)
print(title_case)

uppercase = extract_shape(tokens[0], text.WordShape.IS_UPPERCASE)
print(uppercase)

symbols = extract_shape(tokens[0], text.WordShape.HAS_SOME_PUNCT_OR_SYMBOL)
print(symbols)
```
